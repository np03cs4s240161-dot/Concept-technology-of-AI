```python
#Problem-2
#1
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/medical_students_dataset.csv")
df

```

```python
#2
df.info()
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
#3

#For numerical columns, using mean
nums_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in nums_cols:
  if df[col].isnull().sum() > 0:
    df[col].fillna(df[col].mean(), inplace=True)
    print(f"Filled missing values in '{col}' using MEAN (numeric.column).")

#For categorical columns, using mode
cat_cols=df.select_dtypes(include=['object']).columns
for col in cat_cols:
  if df[col].isnull().sum() > 0:
    df[col].fillna(df[col].mode()[0], inplace= True)
    print(f"Filled missing values in '{col}' using MODE (categorical column).")
```

```python
#4
#Checking how  many duplicate rows exists
print("Number of duplicate rows:", df.duplicated().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)
print("Duplicates removed successfully")
```

```python
#3.2
#Problem-1
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
subset_df = df[['Name','Pclass','Sex','Age','Fare','Survived']]
first_class_df= subset_df[subset_df['Pclass']==1] #Filters only First-class passengers

# Calculating statistics for 'Fare'
fare_mean=first_class_df['Fare'].mean()
fare_median=first_class_df['Fare'].median()
fare_max=first_class_df['Fare'].max()
fare_min=first_class_df['Fare'].min()

#Printing the results
print("Stastics for Fare-Class Passengers:")
print(f"Mean: {fare_mean}")
print(f"Median: {fare_median}")
print(f"Maximum: {fare_max}")
print(f"Minimum: {fare_min}")
```

```python
#Problem-2
#Counting the number of null values in 'Age' coulmn
null_age_count= first_class_df['Age'].isnull().sum()
print(f"Number of null values in 'Age':{null_age_count}")

#Dropping rows where 'Age' is null
first_class_df = first_class_df.dropna(subset=['Age'])

print("\nAfter dropping null values in 'Age':")
print(first_class_df['Age'].isnull().sum())
```

```python
#Problem-4
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
mean_survival = df.groupby("Sex")["Survived"].mean()
print("Mean Survival Rate by Gender:")
print(mean_survival)

#Visualization: bar plot of survival rates
plt.figure(figsize=(6,4))
mean_survival.plot(kind='bar')

plt.title("Mean Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
#Problem-5
import pandas as pd
import matplotlib.pyplot as plt

# 1ï¸âƒ£ Load dataset
df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

# 2ï¸âƒ£ Drop rows where 'Embarked' is missing
df = df.dropna(subset=['Embarked'])

# 3ï¸âƒ£ Calculate mean survival rate for each embarkation port
survival_by_embarked = df.groupby('Embarked')['Survived'].mean()

print("Mean Survival Rate by Embarkation Port:")
print(survival_by_embarked)

# 4ï¸âƒ£ Visualization: bar chart
plt.figure(figsize=(6,4))
survival_by_embarked.plot(kind='bar')

plt.title("Mean Survival Rate by Port of Embarkation")
plt.xlabel("Embarked")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)

plt.show()
```

```python
#Optional Problem-6
import pandas as pd
import matplotlib.pyplot as plt

# 1ï¸âƒ£ Load dataset
df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

# 2ï¸âƒ£ Drop missing age values (important for quantile binning)
df = df.dropna(subset=['Age'])

# 3ï¸âƒ£ Create 5 quantile-based age groups
df['Age_Group'] = pd.qcut(df['Age'], 5, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5'])

# 4ï¸âƒ£ Calculate mean survival rate by Pclass and Age group
survival_table = df.groupby(['Pclass', 'Age_Group'])['Survived'].mean()

print("Mean Survival Rates by Passenger Class and Age Group:")
print(survival_table)

# 5ï¸âƒ£ Visualization
plt.figure(figsize=(10,6))

# Unstack Pclass to create side-by-side bars for each age group
survival_table.unstack(0).plot(kind='bar', figsize=(10,6))

plt.title("Survival Rate by Age Group and Passenger Class")
plt.xlabel("Age Group (Quantiles)")
plt.ylabel("Mean Survival Rate")
plt.xticks(rotation=0)
plt.legend(title="Pclass")

plt.show()
```
